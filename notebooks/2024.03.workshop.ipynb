```python
from qiskit.circuit import QuantumCircuit, QuantumRegister
import matplotlib.pyplot as plt
import crsq_arithmetic as ari
```

```python
dpi=180
qstyle={'name': 'default', 'dpi': dpi, 'fontsize': 24, 'subfontsize': 18}
ratios=[1, 1.2]
fig, axs = plt.subplots(1,2, figsize=(12,4), width_ratios=ratios,dpi=dpi,layout='constrained')

#sadder
n=3
a = QuantumRegister(n, 'a')
b = QuantumRegister(n, 'b')
carry = QuantumRegister(n, 'cr')
qc = QuantumCircuit(a, b, carry)

# to make the internals visible:
qc.append(ari.signed_adder_gate(n), a[:] + b[:] + carry[:])
qc.barrier()
ari.signed_adder(qc, a, b, carry, use_gates=True)

ax=axs[0]
qc.draw(ax=ax,style=qstyle)

#sadder
n=3
a = QuantumRegister(n, 'a')
b = QuantumRegister(n, 'b')
carry = QuantumRegister(n, 'cr')
qc = QuantumCircuit(a, b, carry)

# to make the internals visible:
qc.append(ari.signed_adder_gate(n), a[:] + b[:] + carry[:])
qc.barrier()
ari.signed_adder(qc, a, b, carry, use_gates=False)

ax=axs[1]
qc.draw(ax=ax,style=qstyle)

filename='output/workshop_diagrams/arith-sadder3+scoadder3.1.png'

# save
fig.savefig(filename)
```
